# rnn_gradient_graph

```
import numpy as np
import matplotlib.pyplot as plt

N = 2 # 미니배치 크기
H = 3 # 은닉 상태 벡터의 차원 수
T = 20 # 시계열 데이터의 길이

dh = np.ones((N, H)) # np.ones : 모든 원소가 1인 행렬 반환
np.random.seed(3) # 재현할 수 있도록 난수의 시드 조절
 """기울기 폭발 → 대책 = 기울기 클리핑"""
# Wh = np.random.randn(H, H), 변경전 (시간 비례 증가 그래프) : 신경망 학습 제대로 안됌
 """기울기 소실"""
Wh = np.random.randn(H, H) * 0.5 # 변경후 (시간 비례 감소 그래프) : 기울기가 일정 수준 이하로 작아지면 가중치 매개변수가 더 이상 갱신 안 되서 장기의존 관계 학습 X

norm_list = []
for t in range(T):
    dh = np.matmul(dh, Wh.T)
    norm = np.sqrt(np.sum(dh**2)) / N # norm : 노름(여기선 L2 노름(L2 노름 : 각각의 원소를 제곱해 모두 더하고 제곱근을 취한 값))
    norm_list.append(norm)

plt.plot(norm_list)
plt.show()
```

# 기울기 클리핑, Gradients clipping

```
import numpy as np

dW1 = np.random.rand(3, 3) * 10
dW2 = np.random.rand(3, 3) * 10
grads = [dW1, dW2]
max_norm = 5.0

def clip_grads(grads, max_norm):
    total_norm = 0
    for grad in grads:
        total_norm += np.sum(grad ** 2)
    total_norm = np.sqrt(total_norm)

    rate = max_norm / (total_norm + 1e-6)
    if rate < 1:
        for grad in grads:
            grad *= rate
```
